```python
import cv2
import numpy as np
```

```python
image = cv2.imread("C:/Users/Business Intelli/Desktop/cats vs dogs/train/dogs/dog.109.jpg", 1).astype(np.float32)/255
print('Image Shape:', image.shape)
print('Image Datatype:', image.dtype)
```

```python
#Swap the red and blue channels and display the result:
image[:, :, [0, 2]] = image[:, :, [2, 0]]
cv2.imshow('blue_and_red_swapped', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Swap the channels back and scale them differently to change the colorization of the image:
image[:, :, [0, 2]] = image[:, :, [2, 0]]
image[:, :, 0] = (image[:, :, 0] * 0.9).clip(0, 1) 
image[:, :, 1] = (image[:, :, 1] * 1.1).clip(0, 1)
cv2.imshow('image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
grey = cv2.imread("C:/Users/Business Intelli/Desktop/cats vs dogs/train/dogs/dog.116.jpg", 0)
cv2.imshow('original grey', grey)
cv2.waitKey()
cv2.destroyAllWindows()
print("grey shape:", grey.shape)
print("grey dtype:", grey.dtype)
```

```python
hist, bins = np.histogram(grey, 256, [0, 255])
```

```python
plt.fill(hist)
plt.xlabel('pixel value')
plt.show()
```

```python
grey_eq = cv2.equalizeHist(grey)
```

```python
hist, bins = np.histogram(grey_eq, 256, [0, 255])
plt.fill_between(range(256), hist, 0)
plt.xlabel('pixel value')
plt.show()
```

```python
cv2.imshow('equalized grey', grey_eq)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python

image8 = cv2.imread("C:/Users/Business Intelli/Desktop/cats vs dogs/train/dogs/dog.118.jpg", 1)
cv2.imshow('image', image8)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# image changing to HSV
hsv1  = cv2.cvtColor(image8, cv2.COLOR_BGR2HSV)
cv2.imshow('hsv', hsv1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Equalize the V channel of the HSV image and convert it back to the RGB color space:
hsv1[..., 2] = cv2.equalizeHist(hsv1[..., 2])
color_eq = cv2.cvtColor(hsv1, cv2.COLOR_HSV2BGR)
cv2.imshow('original color', image8)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Show the equalized full color image:
cv2.imshow('equalized color', color_eq)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image7 = cv2.imread('C:/Users/Business Intelli/Desktop/cats vs dogs/train/dogs/dog.117.jpg').astype(np.float32) / 255
print('Shape:', image7.shape)
print('Data type:', image7.dtype)
cv2.imshow('Image', image7)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python

#Convert the image to HSV color space:
hsv = cv2.cvtColor(image7, cv2.COLOR_BGR2HSV)
print('Converted to HSV')
print('Shape:', hsv.shape)
print('Data type:', hsv.dtype)
cv2.imshow('hsv', hsv)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Increase the brightness of the image by multiplying the V channel by some value. Then convert the image to the RGB color space:

hsv[:, :, 2] *= 2
from_hsv = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
print('Converted back to BGR from HSV')
print('Shape:', from_hsv.shape)
print('Data type:', from_hsv.dtype)
cv2.imshow('from_hsv', from_hsv)
cv2.waitKey()
cv2.destroyAllWindows()
```
